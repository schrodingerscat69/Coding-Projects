```python
import matplotlib.pyplot as plt
from math import pi, sin, cos

# Define variables globally
m1 = 2
m2 = 1
theta = pi/4
g = 9.81
t = 0
dt = 0.01
s = 10
x1 = 0
sdot = 0
x1dot = 0

# Initialize arrays to store values for position
x1_arr = []
x2_arr = []
s_arr = []
t_arr = []

# loop for calculating x1ddot and sddot at intervals of t + dt. 
while t < 0.5:
    x1ddot = (-m2 * g * sin(theta) * cos(theta)) / (m1 + m2 * (sin (theta))**2)
    sddot = (m1+m2) * x1ddot / (m2 * cos (theta))
    x1dot += x1ddot * dt
    sdot += sddot * dt
    x1 += x1dot*dt
    x1_arr.append(x1)
    s += sdot*t
    s_arr.append(s)
    t += dt
    t_arr.append(t)
    x2 = x1 - s * cos(theta)
    x2_arr.append(x2)
    
    
# Assuming t_arr, x1_arr, and x2_arr are defined as arrays of data
plt.plot(t_arr, x1_arr, label="Horizontal motion of wedge")
plt.plot(t_arr, x2_arr, label="Horizontal motion of sliding mass")
plt.plot(t_arr, s_arr, label= "Motion down the ramp of m2")

# Adding labels for clarity
plt.xlabel("Time (s)")   # Label for the x-axis
plt.ylabel("Position (m)")  # Label for the y-axis

# Display legend
plt.legend()

# Show the plot
plt.show()

# Momentum of the system should be conserved
print("p1 = ", m1*x1dot)
print("p2 = ", m2*(x1dot-sdot*(cos(theta))))
```
