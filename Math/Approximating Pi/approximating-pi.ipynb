```python
import numpy as np
import matplotlib.pyplot as plt
import time
import matplotlib 
matplotlib.use('TkAgg')


def plot_function():
    fig, axis = plt.subplots(1,2, figsize=(10,5))
    
    # Generated Dots
    axis[0].set_title('Random Generated Dots')
    axis[0].set_aspect('equal')
    axis[0].set_xlim([-1,1])
    axis[0].set_ylim([-1,1])
    
    # Approximation plot
    axis[1].set_title('Approximating Pi')
    axis[1].set_xlim([0,2000])
    axis[1].set_ylim([0,10])
    axis[1].grid()
    
    return axis

axis = plot_function()
plt.ion()

# Number of samples
N_samples = np.arange(1,10000)
pi_counter = 0

xdots = []
ydots = []
dots_color = []
pi_array = []

scatter = axis[0].scatter(xdots, ydots, color=dots_color, marker= 'o', s = 3)
plot, = axis[1].plot([], pi_array, 'black')

for N in N_samples:
    x = np.random.uniform(-1,1)
    y = np.random.uniform(-1,1)
    
    xdots.append(x)
    ydots.append(y)
    
    if np.sqrt(x**2 + y**2) <= 1:
        pi_counter += 1
        dots_color.append('red')
        
    else:
        dots_color.append('blue')
        
    probability = pi_counter / N
    approx_pi = 4*probability
    pi_array.append(approx_pi)
    
    # Updating the plot
    if N%100 == 0:
        scatter.set_offsets(np.column_stack((xdots,ydots)))
        scatter.set_facecolors(dots_color)
        axis[0].set_title(f"Red Dots: {pi_counter}, Blue Dots: {N-pi_counter}")

        plot.set_data(range(1,N+1), pi_array)
        plot.set_color('black')
        
        if N%2000 == 0:
            axis[1].set_xlim([0,N+2000])
            axis[1].set_ylim([approx_pi-1/approx_pi, approx_pi+1/approx_pi])
            axis[1].set_title(f"Approximating Pi: {approx_pi:.5f}")
    
        plt.draw()
        plt.pause(0.01)
    
#print(
   # f"For {N}-samples, Approximated Pi = {approx_pi:.5f},"
   # f"Error : {100*np.abs(approx_pi-np.pi)/np.pi:.3f}%", end='\r')
    

    #time.sleep(0.001)
plt.ioff()
plt.show()
```
