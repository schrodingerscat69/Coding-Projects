```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d_vectors(vectors):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    for v in vectors:
        # Each vector v should be a list or array of 3 elements
        ax.quiver(0, 0, 0, v[0], v[1], v[2])

    # Set labels and title
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    ax.set_title('3D Vector Plot')

    # Set the limits for the axes
    ax.set_xlim([-5, 5])
    ax.set_ylim([-5, 5])
    ax.set_zlim([-5, 5])

    plt.show()

# Example usage
vectors = [[1, 0, 0], [0, 1, 0], [0, 0, 1]] # Standard basis vectors
plot_3d_vectors(vectors)
vector_1 = [[1,0,2],[1, 5, 3],[0,0,-1]]
plot_3d_vectors(vector_1)
```
