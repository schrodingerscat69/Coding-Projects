```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt

def simulate_n_slit_diffraction(N, lambda_light, d, L, a, x_min, x_max, resolution):
    """
    Simulates an N-slit diffraction pattern.

    Parameters:
    N (int): Number of slits
    lambda_light (float): Wavelength of light (in meters)
    d (float): Distance between slits (in meters)
    L (float): Distance to the screen (in meters)
    a (float): Width of each slit (in meters)
    x_min, x_max (float): Range of the screen (in meters)
    resolution (int): Number of points on the screen
    """
    # Screen setup
    x = np.linspace(x_min, x_max, resolution)
    
    # Calculating the angles
    theta = np.arctan(x / L)
    alpha = np.pi * d * np.sin(theta) / lambda_light
    beta = np.pi * a * np.sin(theta) / lambda_light

    # Calculating intensity
    intensity = ((np.sin(beta) / beta)**2) * ((np.sin(N * alpha) / np.sin(alpha))**2)
    intensity_normalized = intensity / np.max(intensity)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, intensity_normalized)
    plt.title(f'{N}-Slit Diffraction Pattern')
    plt.xlabel('Position on Screen (m)')
    plt.ylabel('Normalized Intensity')
    plt.grid(True)
    plt.show()

# Parameters
N = 50  # Number of slits
lambda_light = 10e-9  # Wavelength in meters (500 nm)
d = 20e-6  # Distance between slits in meters
L = 1  # Distance to the screen in meters
a = 10e-6  # Slit width in meters
x_min, x_max = -0.005, 0.005  # Screen range in meters
resolution = 1000

# Running the simulation
simulate_n_slit_diffraction(N, lambda_light, d, L, a, x_min, x_max, resolution)
```
